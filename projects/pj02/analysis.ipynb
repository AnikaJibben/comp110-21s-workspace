# Project 2 DataBank Analysis
## Hypothesis: Higher percentages of children out of school, at primary school age, correlates to a higher number of deaths of ages 5-9 year olds.

```python
import pandas as pd
NO_SCHOOL_CHILDREN: str = "SE.PRM.UNER.ZS"
DEATHS_OF_5_TO_9: str = "SH.DTH.0509"
```

## First I am going to red in the csv data file to a Data Frame

```python
df = pd.read_csv("../../data/databank_education_2018.csv")
```

## Now, I am going to use the shape built in function to display how many rows and columns there are.
### Followed by using the head function to display only the first 10 roes of the DataFrame.

```python
print(df.shape)
df.head(10)
```

## Here, I am establishing a new variable that will display a DataFrame with only the information I am interested in which is the Country, children out of school(SE.PRM.UNER.ZS), and deaths of children aged 5-9(SH.DTH.0509)
### Once again using the head built in function to display only the first 10 rows

```python
select_data = df[["Country", NO_SCHOOL_CHILDREN, DEATHS_OF_5_TO_9]]
select_data.head(10)
```

## Then, I am using the notna built in function to filter out the N/A and NaN in my data. I print my new shape whcih shows a large decrease in my number of rows and use head againa to show the first 10 rows.

```python
filtered_data = select_data[select_data[NO_SCHOOL_CHILDREN].notna() & select_data[DEATHS_OF_5_TO_9].notna()]
print(filtered_data.shape)
filtered_data.head(10)
```

## In this portion of my analysis I organized one of my main variables being analyzed, percentage of children out of school, into descending order from largest to smallest. 
### Using head to display the first 10 rows

```python
descending_no_school = filtered_data.sort_values(by = NO_SCHOOL_CHILDREN, ascending = False)
descending_no_school.head(10)
```

## Using Matplotlib I am able to create a scatter plot between my two constants to help show their correlation

```python
import matplotlib.pyplot as plt
from scipy.stats import linregress

x = filtered_data[NO_SCHOOL_CHILDREN]
y = filtered_data[DEATHS_OF_5_TO_9]
stats = linregress(x, y)
m = stats.slope
b = stats.intercept
plt.scatter(x, y)
plt.plot(x, m * x + b, color = "red")
plt.xlabel("Percentage of Students Not in Primary School")
plt.ylabel("Number of Deaths of Children Aged 5 to 9")
plt.title("Comparison of Kids Not in School to Number of Deaths")
```

## In this part I am finding the correlation between my two indicators, percentage of children not in school to number of deaths of kids aged 5-9

```python
s1 = df[NO_SCHOOL_CHILDREN]
s2 = df[DEATHS_OF_5_TO_9]
s1.corr(s2)
```
